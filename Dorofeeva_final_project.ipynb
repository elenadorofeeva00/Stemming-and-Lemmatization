**Stemming**

```python
import nltk
```

```python
from nltk.stem import PorterStemmer
from nltk.stem.snowball import SnowballStemmer
from nltk.tokenize import sent_tokenize, word_tokenize
```

```python
porter = PorterStemmer()
englishStemmer=SnowballStemmer("english")
```

```python
def stemPorter(sentence):
    token_words=word_tokenize(sentence)
    token_words
    stem_sentence=[]
    for word in token_words:
        stem_sentence.append(porter.stem(word))
        stem_sentence.append(" ")
    return "".join(stem_sentence)
```

```python
def stemSnowball(sentence):
    token_words=word_tokenize(sentence)
    token_words
    stem_sentence=[]
    for word in token_words:
        stem_sentence.append(englishStemmer.stem(word))
        stem_sentence.append(" ")
    return "".join(stem_sentence)
```

```python
for i in range(3):
    print("Sentence: ")
    print(my_lines_list[i], '\n')
    print("Porter stemmed sentence: ")
    print(stemPorter(my_lines_list[i]), '\n')
    print("Snowball stemmed sentence: ")
    print(stemSnowball(my_lines_list[i]), '\n')
```

**Lemmatization**

```python
from nltk.stem import WordNetLemmatizer
from nltk.corpus import wordnet

wordnet_lemmatizer = WordNetLemmatizer()
```

```python
def get_wordnet_pos(word):
    """Map POS tag to first character lemmatize() accepts"""
    tag = nltk.pos_tag([word])[0][1][0].upper()
    tag_dict = {"J": wordnet.ADJ,
                "N": wordnet.NOUN,
                "V": wordnet.VERB,
                "R": wordnet.ADV}
    return tag_dict.get(tag, wordnet.NOUN)
```

```python
punctuations="?:!.,;"
```

```python
print("{0:20}{1:20}".format("Word","Lemma"))

for i in range(3):
    sentence_words = nltk.word_tokenize(my_lines_list[i])
    for word in sentence_words:
        if word in punctuations:
            sentence_words.remove(word)
            
    for word in sentence_words:
        print ("{0:20}{1:20}".format(word,wordnet_lemmatizer.lemmatize(word, get_wordnet_pos(word))))
```

**MultipartiteRank**

```python
grammar="NP: {<ADJ>*<NOUN|PROPN>+}"
```

```python
def evaluate(top_N_keyphrases, references):
    P = len(set(top_N_keyphrases) & set(references)) / len(top_N_keyphrases)
    R = len(set(top_N_keyphrases) & set(references)) / len(references)
    F = (2*P*R)/(P+R) if (P+R) > 0 else 0 
    return (P, R, F)
```

```python
phrases = ['bilbo', 'frodo', 'bag end', 'hobbits', 'hobbiton', 'shire', 'birthday']
```

```python
extractor = pke.unsupervised.MultipartiteRank()
extractor.load_document(input=text, language='en')
extractor.grammar_selection(grammar=grammar)
extractor.candidate_weighting()
keyphrases = extractor.get_n_best(n=7, stemming=True)

top7 = [candidate for candidate, weight in keyphrases]
print("top-7 keyphrases:", '; '.join(top7))

# оценим модель
P, R, F = evaluate(top7, phrases)
print("P@5: {:.3f} R@5: {:.3f} F@5: {:.3f}".format(P, R, F))
```
